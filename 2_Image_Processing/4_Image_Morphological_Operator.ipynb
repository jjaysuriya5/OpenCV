```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros( (600,600) )
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText( blank_img , text = 'SURIYA' , org=(30,350) , fontFace = font , fontScale = 5 , color = (255,255,255)  , thickness = 30 )
print('Creating a BLANK Image')
```

# eroding Image

```python
plt.figure(figsize = (15, 12 ) )
plt.imshow(blank_img , cmap = 'gray')
```

```python
kernel = np.ones( (5,5) , dtype = np.int8 )

result = cv2.erode( blank_img , kernel , iterations = 5 )
```

```python
plt.figure(figsize = (15, 12 ) )
plt.imshow(result , cmap = 'gray')
```

# Noise removal 1

```python
white_noise =  np.random.randint( low = 0 , high = 2 , size = (600,600) )
```

```python
plt.figure(figsize = (15, 8 ) )
plt.imshow(white_noise , cmap = 'gray')
```

```python
white_noise = white_noise * 255 
noise_image = blank_img + white_noise
```

```python
plt.figure(figsize = (15, 8 ) )
plt.imshow(noise_image , cmap = 'gray')
```

```python
clear_image = cv2.morphologyEx( noise_image , cv2.MORPH_OPEN , kernel )

plt.figure(figsize = (15, 8 ) )
plt.imshow(clear_image , cmap = 'gray')
```

# Noise removal 2

```python
white_noise =  np.random.randint( low = 0 , high = 2 , size = (600,600) )
white_noise = white_noise * -255 
noise_image = blank_img + white_noise

noise_image[noise_image==-255] = 0 

plt.figure(figsize = (15, 8 ) )
plt.imshow(noise_image , cmap = 'gray')
```

```python
clear_image = cv2.morphologyEx( noise_image , cv2.MORPH_CLOSE , kernel )

plt.figure(figsize = (15, 8 ) )
plt.imshow(clear_image , cmap = 'gray')
```

# Noise removal 3- Edge Detection

```python
clear_image = cv2.morphologyEx( blank_img , cv2.MORPH_GRADIENT , kernel )

plt.figure(figsize = (15, 8 ) )
plt.imshow(clear_image , cmap = 'gray')
```
