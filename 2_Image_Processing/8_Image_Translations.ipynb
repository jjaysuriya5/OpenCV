```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread( 'jet.jpg' )
img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB) 
```

```python
img.shape
```

```python
plt.imshow( img )
```

# Affine Transformation

```python
height , width = img.shape[:2]
```

```python
quater_height , quater_width = height/4 , width/4 
```

```python
T = np.array( [ [ 1 , 0 , quater_height] , [0,1,0] ] )
img_affine = cv2.warpAffine( img , T ,(width , height ) )
plt.imshow( img_affine )
```

```python
T = np.array( [ [ 1 , 0 , 0] , [0,1,quater_width] ] )
img_affine = cv2.warpAffine( img , T ,(width , height ) )
plt.imshow( img_affine )
```

```python
T = np.array( [ [ 1 , 0 , quater_height] , [0,1,quater_width] ] )
img_affine = cv2.warpAffine( img , T ,(width , height ) )
plt.imshow( img_affine )
```
