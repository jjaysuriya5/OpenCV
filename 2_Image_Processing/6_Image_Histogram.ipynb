```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('red.jpg')
img1 = cv2.cvtColor( img1 , cv2.COLOR_BGR2RGB)

img2 = cv2.imread('green.jpg')
img2 = cv2.cvtColor( img2 , cv2.COLOR_BGR2RGB)

img3 = cv2.imread('blue.jpg')
img3 = cv2.cvtColor( img3 , cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img1 )

plt.subplot(1,2,2)
color = ('b','g','r')
for i,j in enumerate(color):
    hist = cv2.calcHist([img1],channels = [i], mask = None,histSize = [256] , ranges=[0,256])
    plt.plot(hist)
plt.title('Color Histogram')
    
plt.show()
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img2 )

plt.subplot(1,2,2)
color = ('b','g','r')
for i,j in enumerate(color):
    hist = cv2.calcHist([img2],channels = [i], mask = None,histSize = [256] , ranges=[0,256])
    plt.plot(hist)
plt.title('Color Histogram')
    
plt.show()
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img3 )

plt.subplot(1,2,2)
color = ('b','g','r')
for i,j in enumerate(color):
    hist = cv2.calcHist([img3],channels = [i], mask = None,histSize = [256] , ranges=[0,256])
    plt.plot(hist)
plt.title('Color Histogram')
    
plt.show()
```

# Masking and checking the histogram 

```python
mask = np.zeros( img3.shape[:2] , np.uint8 )

plt.imshow( mask , cmap = 'gray' )
```

```python
mask[400:600 , 200 : 600 ] = 255 

plt.imshow( mask , cmap = 'gray' )
```

```python
masked_img = cv2.bitwise_and( img3 , img3 , mask = mask )

plt.imshow( masked_img , cmap = 'gray' )
```

```python
hist = cv2.calcHist([img3],channels = [2], mask = mask ,histSize = [256] , ranges=[0,256])
plt.plot(hist)
```

# Equalization

```python
path = 'C:\\Users\\Administrator\\Desktop\\python\\OpenCV\\1_Image_Processing_Basics'
img = cv2.imread(path+'\\ironman.jpg'  )
img = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img , cmap = 'gray' )

plt.subplot(1,2,2)
plt.title('Equalized Histogram Image with more contrast')
eq_img = cv2.equalizeHist( img )
plt.imshow( eq_img , cmap = 'gray' )
```

- To increase the contrast of the color image , first convert the image to HSV format and then equalize the value channel 

```python
path = 'C:\\Users\\Administrator\\Desktop\\python\\OpenCV\\1_Image_Processing_Basics'
img = cv2.imread(path+'\\ironman.jpg'  )
img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img )


plt.subplot(1,2,2)
img = cv2.cvtColor( img , cv2.COLOR_RGB2HSV )
img[: , : , 2] = cv2.equalizeHist( img[: , : , 2]  )
plt.title('Equalized Histogram Image with more contrast for color image')
img = cv2.cvtColor( img , cv2.COLOR_HSV2RGB )
plt.imshow( img )

plt.show()
```
