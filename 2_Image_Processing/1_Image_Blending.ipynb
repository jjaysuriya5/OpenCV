# Blending the Image 

- Images has to be in same size before resizing 

```python
img1 = cv2.resize( img1 , (5500,3500) ) 
img2 = cv2.resize( img2 , (5500,3500) ) 
```

```python
blended_img = cv2.addWeighted( src1 = img1 , alpha = 0.8 ,src2 = img2 , beta = 0.25 , gamma = 5 )
```

```python
plt.figure(figsize = (10 , 8 ))

plt.title('Blended Image')
plt.imshow( blended_img )

plt.show()
```

# Overlapping Image

```python
img1 = cv2.imread(path+'\\ironman.jpg')
img1 = cv2.cvtColor( img1 , cv2.COLOR_BGR2RGB)

img2 = cv2.imread('avenger.png')
img2 = cv2.cvtColor( img2 , cv2.COLOR_BGR2RGB)
```

```python
large_img = cv2.resize( img2 , (1000,1000) ) 
small_img = cv2.resize( img1 , (200,200) )

x_offset = 0
y_offset = 0 

x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]

large_img[x_offset:x_end , y_offset:y_end ] = small_img
```

```python
plt.figure(figsize = (15 , 8 ))

plt.title('Overlapping Image')
plt.imshow( large_img )

plt.show()
```
