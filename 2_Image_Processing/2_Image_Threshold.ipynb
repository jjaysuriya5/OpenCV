```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
path = 'C:\\Users\\Administrator\\Desktop\\python\\OpenCV\\1_Image_Processing_Basics'

img1 = cv2.imread(path+'\\ironman.jpg')
img1 = cv2.cvtColor( img1 , cv2.COLOR_BGR2GRAY)

plt.figure(figsize = (15 , 8 ))

plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow( img1  , cmap = 'gray' )
```

```python
img1.mean()
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(3,2,1)
plt.title('Orginal Image')
plt.imshow( img1  , cmap = 'gray' )

plt.subplot(3,2,2)
thres , new_img = cv2.threshold( img1 , 80 , 255 , cv2.THRESH_TRUNC)
plt.title('THRESH_TRUNC Image')
plt.imshow( new_img , cmap = 'gray')

plt.subplot(3,2,3)
thres , new_img = cv2.threshold( img1 , 80 , 255 , cv2.THRESH_BINARY )
plt.title('THRESH_BINARY Image')
plt.imshow( new_img , cmap = 'gray')

plt.subplot(3,2,4)
thres , new_img = cv2.threshold( img1 , 80 , 255 , cv2.THRESH_BINARY_INV )
plt.title('THRESH_BINARY_INV Image')
plt.imshow( new_img , cmap = 'gray')

plt.subplot(3,2,5)
thres , new_img = cv2.threshold( img1 , 80 , 255 , cv2.THRESH_TOZERO )
plt.title('THRESH_TOZERO Image')
plt.imshow( new_img , cmap = 'gray')

plt.subplot(3,2,6)
thres , new_img = cv2.threshold( img1 , 80 , 255 , cv2.THRESH_TOZERO_INV )
plt.title('THRESH_TOZERO_INV Image')
plt.imshow( new_img , cmap = 'gray')

plt.tight_layout()
plt.show()
```

# Adaptive Thresholding And Blending 

```python
img_new = cv2.adaptiveThreshold( img1 , 255 , cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY , 15 , 8)
plt.imshow( img_new , cmap = 'gray' )
```

```python
img_new = cv2.adaptiveThreshold( img1 , 255 , cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY , 15 , 8)
plt.imshow( img_new , cmap = 'gray' )
```

```python
img_blnd = cv2.addWeighted( src1 = img1 , alpha = 0.8 , src2 = img_new , beta = 0.2 , gamma = 2 )
plt.imshow(img_blnd , cmap = 'gray')
```
