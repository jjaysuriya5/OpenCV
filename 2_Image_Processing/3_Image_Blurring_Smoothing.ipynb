# Kernel Meethod 

```python
img1 = cv2.imread(path+'\\ironman.jpg').astype(np.float32) / 255
img1 = cv2.cvtColor( img1 , cv2.COLOR_BGR2RGB)
img1 = cv2.resize( img1 , (1000,1000) )

font = cv2.FONT_HERSHEY_COMPLEX

cv2.putText( img1 , text = 'HERO' , org = (75,600) , fontFace = font , fontScale = 10 , color = (0,255,0) , thickness = 4)

plt.figure(figsize = (15 , 8 ))
plt.imshow( img1 )
plt.show()
```

```python
kernel = np.ones( (5,5) , dtype = np.float32 ) / 25 

dest = cv2.filter2D( img1 , -1 , kernel )


plt.figure(figsize = (15 , 8 ))
plt.imshow( dest )
plt.show()
```

```python

dest = cv2.blur( img1 , (10,5) )


plt.figure(figsize = (15 , 8 ))
plt.imshow( dest )
plt.show()
```

```python

dest = cv2.GaussianBlur( img1 , (5,5) , 10)


plt.figure(figsize = (15 , 8 ))
plt.imshow( dest )
plt.show()
```

```python

dest = cv2.medianBlur( img1 , 5 )


plt.figure(figsize = (15 , 8 ))
plt.imshow( dest )
plt.show()
```

```python

dest = cv2.bilateralFilter( img1 , 9 , 75 , 75  )


plt.figure(figsize = (15 , 8 ))
plt.imshow( dest )
plt.show()
```
