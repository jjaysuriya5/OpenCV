```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img  = cv2.imread('hello.jpg')
img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )

plt.figure(figsize = (10,8))
plt.imshow( img )
```

```python
# finding the corner points
corner_points = np.float32( [ [25,100] , [160,40] ,[77,185] , [205,115] ] )
# A4 page ratio
page_ratio = np.float32( [ [0,0] ,[420,0] ,[0,594] ,[420,594] ] )

M = cv2.getPerspectiveTransform( corner_points , page_ratio )

wraped = cv2.warpPerspective( img , M , (420 , 594) )

plt.figure(figsize = (10,8))
plt.imshow( wraped )
```
