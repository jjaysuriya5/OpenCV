```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread( 'sudoku.png' , 0 )

plt.figure(figsize = (10, 8 ))
plt.imshow(img , cmap = 'gray')
```

# Edge Detection using threshold and Morplology operation

```python
plt.figure(figsize = (15, 8 ) )

kernel = np.ones((5,5) , dtype = np.int8 )
plt.subplot(1,2,1)
new_img = cv2.morphologyEx( img , cv2.MORPH_GRADIENT , kernel )
plt.title('MORPH_GRADIENT')
plt.imshow(new_img , cmap = 'gray')

plt.subplot(1,2,2)
_ , new_img = cv2.threshold( img , 80 , 255 , cv2.THRESH_BINARY , kernel )
plt.title('THRESH_BINARY')
plt.imshow(new_img , cmap = 'gray')

plt.show()
```

# Image Gradient for edge Detection

```python
plt.figure(figsize = (15, 8 ) )


plt.subplot(1,2,1)
sobel_x = cv2.Sobel(img , cv2.CV_64F , 1 , 0 , ksize = 5 )
plt.title('Sobel along x axis , vertical lines are detected ')
plt.imshow(sobel_x , cmap = 'gray')

plt.subplot(1,2,2)
sobel_y = cv2.Sobel(img , cv2.CV_64F , 0 , 1, ksize = 5 )
plt.title('Sobel along y axis , horizontal lines are detected ')
plt.imshow(sobel_y , cmap = 'gray')

plt.show()
```

```python
# combining sobel_x and sobel_y 

combine_sobel = cv2.addWeighted( src1 = sobel_x , alpha = 0.5 , src2 = sobel_y , beta = 0.5 , gamma = 1 )

plt.figure(figsize = (15, 8 ) )
plt.title('Combining Sobel along x axis and Y axis , edges are detected ')
plt.imshow(combine_sobel , cmap = 'gray')

plt.show()
```

- Combining techniqes helps in identifying the edge better
