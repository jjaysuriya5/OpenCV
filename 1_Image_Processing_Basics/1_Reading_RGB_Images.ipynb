```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
```

```python
color_img = Image.open('intro.png')

plt.figure(figsize=(10,8))

plt.subplot( 1 , 2 , 1 )
plt.title('Orginal Image')
plt.imshow(color_img)

plt.subplot( 1 , 2 , 2 )
plt.title('numpy Image')
plt.imshow(np.array( color_img ))

plt.show()
```

```python
color_img_arr = np.array( color_img )
```

# Color Images has three chanel lets see how the image is displayed

```python
plt.figure(figsize=(10,8))

plt.subplot( 2 , 1 , 1 )
plt.title('Orginal Imamge')
plt.imshow(color_img_arr)

plt.subplot( 2 , 3,  4 )
plt.title('RED Channel of the Image')
plt.imshow(color_img_arr[:,:,0] , cmap = 'gray')

plt.subplot( 2 , 3 , 5 )
plt.title('GREEN Channel of the Image')
plt.imshow(color_img_arr[:,:,1] , cmap = 'gray')

plt.subplot( 2 , 3 , 6)
plt.title('BLUE Channel of the Image')
plt.imshow(color_img_arr[:,:,2] , cmap = 'gray')

plt.show() 
```

- The Corresponding channels are displayes as white image and other color are displated as black image 

# Keeping only one channel value and updating the other channels to zero

```python
red_img = color_img_arr.copy()
blue_img = color_img_arr.copy()
green_img = color_img_arr.copy()

red_img[:,:,[1,2]] = 0  # Updating other channel(green , Blue) values to zero
green_img[:,:,[0,2]] = 0  # Updating other channel(Red , Blue) values to zero
blue_img[:,:,[0,1]] = 0  # Updating other channel(green , Red) values to zero
```

```python
plt.figure(figsize=(10,8))

plt.subplot( 2 , 1 , 1 )
plt.title('Orginal Imamge')
plt.imshow(color_img_arr)

plt.subplot( 2 , 3,  4 )
plt.title('RED Channel of the Image')
plt.imshow(red_img)

plt.subplot( 2 , 3 , 5 )
plt.title('GREEN Channel of the Image')
plt.imshow(green_img)

plt.subplot( 2 , 3 , 6)
plt.title('BLUE Channel of the Image')
plt.imshow(blue_img)

plt.show() 
```

- Only the selected channel is displayed
