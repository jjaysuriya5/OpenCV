```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ironman.jpg')
```

```python
type(img)  # the images is stored as numpy array
```

```python
img.shape
```

```python
plt.figure(figsize = (15,8))

plt.subplot(1,2,1)
plt.title('Reading Using Matplotlib')
plt.imshow( plt.imread('ironman.jpg'))

plt.subplot(1,2,2)
plt.title('Reading Using OpenCV')
plt.imshow( img )

plt.show()
```

- Default Channel order for OpenCV is BGR

```python
cvt_img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
```

```python
plt.figure(figsize = (15,8))

plt.subplot(1,2,1)
plt.title('Reading Using OpenCV')
plt.imshow( img )

plt.subplot(1,2,2)
plt.title('Reading Using OpenCV and converting from BGR to RGB channel')
plt.imshow( cvt_img )

plt.show()
```

- Orginal image is been reproduced

# Resizing images

```python
plt.figure(figsize = (15,8))

plt.subplot(1,3,1)
plt.title('Orginal Image')
plt.imshow( cvt_img )

plt.subplot(1,3,2)
plt.title('Resizing image to given dimension')
plt.imshow(  cv2.resize( cvt_img , (1000 , 1000 ) )   )

plt.subplot(1,3,3)
w_ratio = 0.5
h_ratio = 0.5
plt.title('Resizing image to given ratio')
plt.imshow(  cv2.resize( cvt_img , (0 , 0 ) , cvt_img , w_ratio , h_ratio)  )

plt.show()
```

# Flipping image

```python
plt.figure(figsize = (15,8))

plt.subplot(2,1,1)
plt.title('Orginal Image')
plt.imshow( cvt_img )

plt.subplot(2,3,4)
plt.title('Flipping along X-axis')
plt.imshow( cv2.flip(cvt_img,0) )

plt.subplot(2,3,5)
plt.title('Flipping along Y-axis')
plt.imshow( cv2.flip(cvt_img,1) )

plt.subplot(2,3,6)
plt.title('Flipping along both the axis')
plt.imshow( cv2.flip(cvt_img,-1) )

plt.show()
```

# Saving Image

```python
cv2.imwrite( 'New_IronMan.jpg' , cv2.flip(cvt_img,1)  )
```

```python
plt.imshow( plt.imread('New_IronMan.jpg') )
```
