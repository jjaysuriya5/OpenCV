```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('ironman.jpg')
img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )

plt.figure(figsize = (15 , 8 ))

plt.title('Orginal Image')
plt.imshow( img )
```

```python
edges = cv2.Canny( img , threshold1 = 300 , threshold2 = 300 )

plt.figure(figsize = (15 , 8 ))

plt.title('Edge Image')
plt.imshow( edges )
```

```python
edges = cv2.Canny( img , threshold1 = 0 , threshold2 = 255 )

plt.figure(figsize = (15 , 8 ))

plt.title('Edge Image')
plt.imshow( edges )
```

```python
# better Sugested Formula 

img_med = np.median( img )
lower = int( max ( 0 , 0.7*img_med ) )
upper = int( max ( 255 , 1.3* img_med ))
```

```python
edges = cv2.Canny( img , threshold1 = lower , threshold2 = upper + 150  )

plt.figure(figsize = (15 , 8 ))

plt.title('Edge Image')
plt.imshow( edges )
```

```python
edges = cv2.Canny( img , threshold1 = 0 , threshold2 = 50 )

plt.figure(figsize = (15 , 8 ))

plt.title('Edge Image')
plt.imshow( edges )
```

# Try the same with blurring image

```python
blur_img = cv2.blur( img , ksize =  (10,5) )

edges = cv2.Canny( blur_img , threshold1 = 0 , threshold2 = 15 )

plt.figure(figsize = (15 , 8 ))

plt.title('Edge Image')
plt.imshow( edges )
```
