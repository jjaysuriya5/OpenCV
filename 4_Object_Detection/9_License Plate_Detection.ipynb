# Car Plate detection

```python
img = cv2.imread('car_plate2.jpg')
img = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )

plt.figure(figsize=(15,8))
plt.imshow(img , cmap = 'gray' )
```

```python
harr = 'C:\\Users\\Administrator\\Desktop\\python\\OpenCV\\haar-cascade-files-master'

cascade_path = harr + '\\haarcascade_russian_plate_number.xml'

licence_plate_detection = cv2.CascadeClassifier(cascade_path)
```

```python
def detect_licence_plate( img ):
    licence_plate_img = img.copy()
    
    licence_plate_rect = licence_plate_detection.detectMultiScale(licence_plate_img , scaleFactor = 1.2 , minNeighbors = 7  )
    
    for x , y , w , h in licence_plate_rect :
        cv2.rectangle( licence_plate_img , (x,y) , (x+w , y+h) , (255,255,255) , 10 )
        
    return licence_plate_img
```

```python
result = detect_licence_plate( img )

plt.figure( figsize = (15 , 8) )

plt.subplot( 1 , 2 , 1 )
plt.title('Iron Man')
plt.imshow( img , cmap = 'gray')

plt.subplot( 1 , 2 , 2 )
plt.title('Iron Man with face detection')
plt.imshow( result , cmap = 'gray')
```

```python
licence_plate_rect = licence_plate_detection.detectMultiScale(img , scaleFactor = 1.2 , minNeighbors = 2  )

for x , y , w , h in licence_plate_rect :
    print( x , y , w , h )
```
