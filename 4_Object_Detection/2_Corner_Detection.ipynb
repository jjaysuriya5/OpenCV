```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('chess1.jpg')
blank_chess = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
blank_chess_gray = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )

img = cv2.imread('chess2.jpg')
chess = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
chess_gray = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )
```

```python
plt.figure(figsize = (15 , 8 ))

plt.subplot(2,2,1)
plt.title('Orginal Image')
plt.imshow( blank_chess  )

plt.subplot(2,2,2)
plt.title('GRAY Scale Image' )
plt.imshow( blank_chess_gray  , cmap = 'gray')

plt.subplot(2,2,3)
plt.title('Orginal Image')
plt.imshow( chess  )

plt.subplot(2,2,4 )
plt.title('GRAY Scale Image' )
plt.imshow( chess_gray  , cmap = 'gray')

plt.show()
```

# Corner detection using Harish corner 

```python
dst = cv2.cornerHarris( src = blank_chess_gray , blockSize = 2 , ksize = 3 , k = 0.04 )
dst = cv2.dilate( dst , None )
blank_chess[dst > 0.01*dst.max() ] = [255,0,0 ]

plt.figure(figsize = (15 , 8 ))
plt.imshow( blank_chess )
```

```python
dst = cv2.cornerHarris( src = chess_gray , blockSize = 2 , ksize = 3 , k = 0.04 )
dst = cv2.dilate( dst , None )
chess[dst > 0.01*dst.max() ] = [255,0,0 ]

plt.figure(figsize = (15 , 8 ))
plt.imshow( chess )
```

# Corner detection using shi tomasi corner detection( good feature to track )

```python
img = cv2.imread('chess1.jpg')
blank_chess = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
blank_chess_gray = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )

img = cv2.imread('chess2.jpg')
chess = cv2.cvtColor( img , cv2.COLOR_BGR2RGB )
chess_gray = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )
```

```python
corners = cv2.goodFeaturesToTrack( blank_chess_gray , maxCorners = 100 , qualityLevel = 0.01 , minDistance = 10 )

corners = np.int0( corners )

for i in corners :
    x , y = i.flatten()
    cv2.circle( blank_chess , (x,y) , 5 , (255,0,0) , -1 )

plt.figure(figsize = (15 , 8 ))
plt.imshow( blank_chess )
```

```python
corners = cv2.goodFeaturesToTrack( chess_gray , maxCorners = 40 , qualityLevel = 0.01 , minDistance = 10 )

corners = np.int0( corners )

for i in corners :
    x , y = i.flatten()
    cv2.circle( chess , (x,y) , 3 , (255,0,0) , -1 )

plt.figure(figsize = (15 , 8 ))
plt.imshow( chess )
```
