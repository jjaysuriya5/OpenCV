```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('internal_external.png')
img = cv2.cvtColor( img , cv2.COLOR_BGR2GRAY )

plt.figure(figsize = (15 , 8 ))

plt.title('Orginal Image' )
plt.imshow( img , cmap = 'gray' )
```

```python
contours , hierarchy =  cv2.findContours( img , cv2.RETR_CCOMP , cv2.CHAIN_APPROX_SIMPLE )  
```

# external Contour

```python
 

external_countour = np.zeros( img.shape )

for i in range(len(contours)) :
    if hierarchy[0][i][3] == -1 :
        cv2.drawContours( external_countour , contours , i , 255 , -1  )
        
plt.imshow(external_countour , cmap = 'gray')
```

# internal Contour

```python


internal_countour = np.zeros( img.shape )

for i in range(len(contours)) :
    if hierarchy[0][i][3] != -1 :
        cv2.drawContours( internal_countour , contours , i , 255 , -1  )
        
plt.imshow(internal_countour , cmap = 'gray')
```
